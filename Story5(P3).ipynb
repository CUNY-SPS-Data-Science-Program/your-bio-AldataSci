```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly as px
import seaborn as sns
```

```python
Temp = pd.read_csv('https://raw.githubusercontent.com/CUNY-SPS-Data-Science-Program/your-bio-AldataSci/main/Story%205/temperature_fig-1%20(1).csv')

Temp.head(5)
```

```python
Temp2 = Temp[Temp['Year'] >= 1950]
```

```python
fig,ax = plt.subplots(figsize=(16, 9))
ax.xaxis.set_tick_params(pad=10)
ax.yaxis.set_tick_params(pad=10)

# Create an inverted bar chart. with negative values going down and positive values going up..

# add a line grid..
ax.grid(color='black',axis = 'y',
        linestyle='-.', linewidth=0.4,
        alpha=0.8)
plt.bar(Temp2['Year'][Temp2['Earth\'s surface'] < 0],Temp2['Earth\'s surface'][Temp2['Earth\'s surface'] < 0],color = 'b')
plt.bar(Temp2['Year'][Temp2['Earth\'s surface'] > 0],Temp2['Earth\'s surface'][Temp2['Earth\'s surface'] > 0], color = 'r')
plt.suptitle('Average Global Temperature', ha = 'left', x = 0.126, y = 0.96, fontsize = 14, fontweight = 'bold')
ax.set_title('The Century Heats up: A Climate Turning Point at the Turn of the Millennium',
             loc='left',fontsize = 10 )
plt.xlabel('Year')
plt.ylabel('Temperature (In Celsius)')
plt.show()
```

```python
## Take a look at the hurricane data listing from The Hurricane Research Division..

Hurricane = pd.read_csv('https://raw.githubusercontent.com/CUNY-SPS-Data-Science-Program/your-bio-AldataSci/main/Story%205/cyclones_fig-1.csv')

H2 = Hurricane[Hurricane['Year'] >= 1950]

H2.head(5)
```

```python
H2['Year'].unique()
```

```python
## Let utilize a seaborn plot to make the line graph better..

fig,ax = plt.subplots(figsize=(16, 9))
ax.grid(color='black',axis = 'y',
        linestyle='--', linewidth=0.4,
        alpha=0.8)
plt.plot(H2['Year'],H2['Total hurricanes (adjusted)'],color = 'b')
plt.plot(H2['Year'],H2['Hurricanes reaching the United States'],color = 'r')
plt.suptitle('Number of Hurricanes in The North Atlantic (1940-2018)', ha = 'left', x = 0.126, y = 0.96, fontsize = 14, fontweight = 'bold')
ax.set_title('Decades of Disturbances: Tracking Hurricane Occurrences Over Time',
             loc='left',fontsize = 10 )
plt.xlabel('Year')
plt.ylabel('Number of Hurricane Occurrences')
plt.legend(["Total Hurricanes","Hurricanes reaching the US"])
plt.show()
```

```python
## This is the sea surface temperature, warmer sea temperature means stronger and frequent hurricane

Sealvl = pd.read_csv('https://raw.githubusercontent.com/CUNY-SPS-Data-Science-Program/your-bio-AldataSci/main/Story%205/sea-surface-temp_fig-1.csv')


Sealvl2 = Sealvl[Sealvl['Year'] >= 1950]
## Use the default style to prevent the seaborn style from fitting in..
plt.style.use('default')
fig,ax = plt.subplots(figsize = (15,8))
ax.grid(color='black',axis = 'y',
        linestyle='-.', linewidth=0.4,
        alpha=0.8)
plt.suptitle('Average Global Sea Temperature (1950-2020)', ha = 'left', x = 0.126, y = 0.96, fontsize = 14, fontweight = 'bold')
ax.set_title('Surging Heat: A Rise in Sea Temperature from the 70s to the present',
             loc='left',fontsize = 10 )
plt.xlabel('Year')
plt.ylabel('Temperature (In Farenheit)')

## Add The Text on the Line
plt.text(Sealvl2['Year'].iloc[2],Sealvl2['Annual anomaly'].iloc[29],'1970-2020 Average Temperature Baseline',fontsize = 10)
plt.axhline(y = 0,color = 'black',linestyle = "--",label = "reference line")
plt.plot(Sealvl2['Year'],Sealvl2['Annual anomaly'],color = 'red')
plt.show()
```

```python
## ## Check the sea levels if the sea levels are increasing then that means there are going to more and frequent hurricanes..

SeaLevel = pd.read_csv('https://raw.githubusercontent.com/CUNY-SPS-Data-Science-Program/your-bio-AldataSci/main/Story%205/sea-level_fig-1.csv')

## Place the data after 1950
SeaLevel2 = SeaLevel[SeaLevel['Year'] >= 1950]

## Create The Chart:
fig,ax = plt.subplots(figsize = (15,8))
ax.grid(color='black',axis = 'y',
        linestyle='-', linewidth=0.4,
        alpha=0.8)
plt.plot(SeaLevel2['Year'],SeaLevel2['CSIRO - Adjusted sea level (inches)'], color = "blue")
plt.plot(SeaLevel2['Year'],SeaLevel2['NOAA - Adjusted sea level (inches)'],color = "red")
## set the axes to zero.
ax.set_ylim(ymin=0)
plt.suptitle('Average Sea Level Over Time (1950-2020)', ha = 'left', x = 0.126, y = 0.96, fontsize = 14, fontweight = 'bold')
ax.set_title('The Relentless Tide: Escalating Sea Levels',
             loc='left',fontsize = 10 )
plt.xlabel('Year')
plt.ylabel('Adjusted Sea Levels (In Inches)')
plt.legend(['Tide-Gauge Measurement','Satellite Measurement'])

# Placeholder this is the code to put in plt.legend ,bbox_to_anchor=(1.05, 1), loc='upper left')
```
